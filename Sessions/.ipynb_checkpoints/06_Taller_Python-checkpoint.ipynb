##  <a id='t14'>14. ¿QUÉ ES GITHUB?</a>

GitHub es una plataforma de desarrollo colaborativo basada en la nube que permite a los desarrolladores alojar y revisar código, gestionar proyectos y colaborar en software. GitHub utiliza el sistema de control de versiones Git, que facilita el seguimiento de los cambios en el código y la coordinación entre múltiples desarrolladores que trabajan en un mismo proyecto.

https://github.com/skills/introduction-to-github?tab=readme-ov-file

##  <a id='t14.1.'>14.1 ¿Para qué sirve?</a>

**Control de versiones:** Permite a los desarrolladores seguir y gestionar cambios en el código, facilitando la colaboración y la corrección de errores.

**Colaboración:** Los desarrolladores pueden trabajar juntos en proyectos desde cualquier parte del mundo. Pueden revisar el código de otros, discutir cambios y contribuir con sus propias modificaciones.

**Alojamiento de repositorios:** GitHub ofrece almacenamiento para los proyectos, permitiendo a los desarrolladores alojar sus repositorios de código de forma segura y accesible.

**Documentación:** Los proyectos en GitHub pueden incluir documentación detallada, instrucciones de uso y guías para los colaboradores.

**Gestión de proyectos:** Herramientas como Issues, proyectos y wikis permiten a los equipos gestionar tareas, reportar errores y planificar el desarrollo del software.

**Integración continua y despliegue continuo (CI/CD):** GitHub se integra con múltiples herramientas y servicios que automatizan las pruebas y el despliegue del código, mejorando la calidad y velocidad del desarrollo.

**Comunidades y redes:** GitHub es una plataforma social donde los desarrolladores pueden seguir a otros, contribuir a proyectos de código abierto y construir una red profesional.

##  <a id='t14.2.'>14.2 ¿Qué es un repositorio?</a>
Un repositorio en GitHub es una ubicación central donde se almacena y gestiona el código fuente de un proyecto. Los repositorios facilitan la colaboración, el control de versiones y el seguimiento de cambios en el código.

**Almacenamiento de código:** Guarda archivos y el historial del proyecto, accesible desde cualquier lugar.

**Control de versiones:** Registra cambios en el código, permitiendo revertir y entender modificaciones.

**Colaboración:** Facilita el trabajo simultáneo de múltiples desarrolladores mediante ramas, commits y pull requests.

**Documentación:** Incluye archivos README, wikis y otros documentos explicativos.

**Gestión de cambios:** Permite proponer, revisar y aprobar cambios antes de integrarlos al código principal.

**Seguimiento de issues:** Rastrea bugs, nuevas características y tareas.

**Automatización:** Se integra con CI/CD para pruebas y despliegues automáticos, mejorando la eficiencia.

**Forking y contribuciones:** Permite crear copias para experimentar y enviar mejoras al proyecto original.
